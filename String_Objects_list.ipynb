```python
import math

n = 9
mid_val = math.ceil(n/2)

for i in range(n):
  if i < n/2:
    for j in range(i+1):
      print('*', end=' ')
  else:
    for k in range(mid_val -(i%mid_val)-1):
      print('*', end=' ')
  print()
```
